## Surface Depth (m)

```python
import pygeos
import rtree

# Variable you are looking for
variable = 'SURFACE_Depth_m'

####
new_df = geopandas.GeoDataFrame(data_new,
                             geometry=geopandas.points_from_xy(data_new.Longitude,
                                                               data_new.Latitude),
                             crs = "+proj=sinu +lon_0=70 +x_0=0 +y_0=0 +a=6371007.181 +b=6371007.181 +units=m +no_defs")

# New dataframe for new variable
temp_variable = new_df[[variable , 'geometry']]

# Drop rows where geometry is empty
temp_variable = temp_variable.loc[temp_variable['geometry'].is_valid, :]



# total area for the grid
xmin, ymin, xmax, ymax= temp_variable.total_bounds
# how many cells across and down
n_cells=100
cell_size = (xmax-xmin)/n_cells
# projection of the grid
crs = "+proj=sinu +lon_0=70 +x_0=0 +y_0=0 +a=6371007.181 +b=6371007.181 +units=m +no_defs"
# create the cells in a loop
grid_cells = []
for x0 in np.arange(xmin, xmax+cell_size, cell_size ):
    for y0 in np.arange(ymin, ymax+cell_size, cell_size):
          # bounds
          x1 = x0-cell_size
          y1 = y0+cell_size
          grid_cells.append( shapely.geometry.box(x0, y0, x1, y1)  )
cell = geopandas.GeoDataFrame(grid_cells, columns=['geometry'], 
                                  crs=crs)
merged = geopandas.sjoin(temp_variable,cell,how ='left',op='within')

# make a simple count variable that we can sum
merged['element_cell']=1
# Compute stats per grid cell -- aggregate fires to grid cells with dissolve
dissolve = merged.dissolve(by="index_right", aggfunc="count")
# put this into cell
cell.loc[dissolve.index, 'element_cell'] = dissolve.element_cell.values

# New dataframe to plot 
temp_variable_2 = pd.DataFrame(data_new[['Longitude',"Latitude"]])
temp_variable_2[variable] = temp_variable[variable]
temp_variable_2 = np.array(temp_variable_2)

x = temp_variable_2[:,0]
y = temp_variable_2[:,1]
z = temp_variable_2[:,2] 
```

```python
# Create grid values first.
xi = np.linspace(x.min(),x.max(),50 )
yi = np.linspace(y.min(),y.max(),50 )

triang = tri.Triangulation(x,y)

interpolator = tri.LinearTriInterpolator(triang, z)
Xi, Yi = np.meshgrid(xi, yi)
zi = interpolator(Xi, Yi)

zi.shape
#plt.imshow(zi)

fig, (ax1) = plt.subplots(nrows=1,figsize=(25,15))
ax1.contour(xi, yi, zi, levels=20, linewidths=0.5, colors='k')
cntr1 = ax1.contourf(xi, yi, zi, levels=20, cmap="RdBu_r")

fig.colorbar(cntr1, ax=ax1)
ax1.plot(x,y, 'ko',mfc='none', ms=5)
ax1.set_title(variable +' 2018-2019')
plt.xticks(rotation=45)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
```

```python
# total area for the grid
xmin, ymin, xmax, ymax= temp_variable.total_bounds
# how many cells across and down
n_cells=100
cell_size = (xmax-xmin)/n_cells
# projection of the grid
crs = "+proj=sinu +lon_0=70 +x_0=0 +y_0=0 +a=6371007.181 +b=6371007.181 +units=m +no_defs"
# create the cells in a loop
grid_cells = []
for x0 in np.arange(xmin, xmax+cell_size, cell_size ):
    for y0 in np.arange(ymin, ymax+cell_size, cell_size):
          # bounds
          x1 = x0-cell_size
          y1 = y0+cell_size
          grid_cells.append( shapely.geometry.box(x0, y0, x1, y1)  )
cell = geopandas.GeoDataFrame(grid_cells, columns=['geometry'], 
                                  crs=crs)
merged = geopandas.sjoin(temp_variable,cell,how ='left',op='within')

# make a simple count variable that we can sum
merged['element_cell']=1
# Compute stats per grid cell -- aggregate fires to grid cells with dissolve
dissolve = merged.dissolve(by="index_right", aggfunc="count")
# put this into cell
cell.loc[dissolve.index, 'element_cell'] = dissolve.element_cell.values

# New dataframe to plot 
temp_variable_2 = pd.DataFrame(data_new[['Longitude',"Latitude"]])
temp_variable_2['SURFACE_POC_mg/L'] = temp_variable['SURFACE_POC_mg/L']
temp_variable_2 = np.array(temp_variable_2)

x = temp_variable_2[:,0]
y = temp_variable_2[:,1]
z = temp_variable_2[:,2] 

# Create grid values first.
xi = np.linspace(x.min(),x.max(),50 )
yi = np.linspace(y.min(),y.max(),50 )

triang = tri.Triangulation(x,y)

interpolator = tri.LinearTriInterpolator(triang, z)
Xi, Yi = np.meshgrid(xi, yi)
zi = interpolator(Xi, Yi)

zi.shape
#plt.imshow(zi)

fig, (ax1) = plt.subplots(nrows=1,figsize=(25,15))
ax1.contour(xi, yi, zi, levels=20, linewidths=0.5, colors='k')
cntr1 = ax1.contourf(xi, yi, zi, levels=20, cmap="RdBu_r")

fig.colorbar(cntr1, ax=ax1)
ax1.plot(x,y, 'ko',mfc='none', ms=5)
ax1.set_title('SURFACE_POC_mg/L 2018-2019 with CV <=0.05')
plt.xticks(rotation=45)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
```
