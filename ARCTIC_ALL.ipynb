```python
# ARCTIC ALL DATAFRAME

# Create new dataframe for BR 2019 data with important columns
BR_2019_data = pd.DataFrame(BR_2019, columns = ['Chlorophyll ug/L',
                                              'fDOM QSU',
                                              'Sal psu',
                                              'pH',
                                              'DOC avg'])

# Join another BR dataframe to BR 2019 data
BR_2019_data = BR_2019_data.add(BR_2019_2,fill_value=0)
BR_2019_data = BR_2019_data.drop(columns =  ['DOC Replicate 1 (mM)',
 'DOC Replicate 2 (mM)'])

# Add column for place ( make it a float)
BR_2019_data['Place'] = 'Barrow'
BR_2019_data['Place'] = BR_2019_data.Place.astype(str)

BR_2019_data

```

```python
# Create new dataframe for PB 2021 data with important columns
PB_2021_data = pd.DataFrame(PB_2021, columns = ['SURFACE_Latitude',
 'SURFACE_Longitude','SURFACE_Chlorophyll_ug/L','SURFACE_fDOM_QSU','DOC_avg','SURFACE_Salinity_psu', 'SURFACE_pH'])

# Add column for place
PB_2021_data['Place'] = 'Prudhoe_Bay'
PB_2021_data['Place'] = PB_2021_data.Place.astype(str)

# Rename the columns to later combine dataframes 
PB_2021_data = PB_2021_data.rename(columns = {'SURFACE_Longitude':'Longitude ', 
                              'SURFACE_Latitude': 'Latitude',
                              'SURFACE_Chlorophyll_ug/L':'Chlorophyll ug/L',
                              'DOC_avg':'DOC avg',
                              'SURFACE_fDOM_QSU':'fDOM QSU',
                              'SURFACE_Salinity_psu':'Sal psu',
                              'SURFACE_pH':'pH'})

PB_2021_data



# BR_2019
# PB_2021
# YK_2019

```

```python

## YK

# Create new dataframe for YK 2019 data with important columns
YK_2019_data = pd.DataFrame(YK_2019, columns = ['latitude (N)',
 'longitude (W)','Chl ug/l CTD','FDOM CTD','salinity CTD','DOC'])

# Add column for place
YK_2019_data['Place'] = 'Yukon'
YK_2019_data['Place'] = YK_2019_data.Place.astype(str)

#temp_2['Place'] = '3'
YK_2019_data = YK_2019_data.dropna()
YK_2019_data = YK_2019_data.rename(columns = {'latitude (N)':'Latitude',
                                 'longitude (W)':'Longitude ',
                                 'Chl ug/l CTD':'Chlorophyll ug/L',
                                 'FDOM CTD':'fDOM QSU',
                                 'salinity CTD':'Sal psu',
                                 'DOC':'DOC avg'})

YK_2019_data
```

```python
# For all places create a new dataframe with all Arctic Data
places = [BR_2019_data, PB_2021_data, YK_2019_data]

ARCTIC_ALL = pd.concat(places)
ARCTIC_ALL 
```

## Edits

```python
# Edits

###
# Fix DOC for all Places -> change to mg/L

#1. YK_2019_data
YK_2019_data['DOC avg'] = YK_2019_data['DOC avg']/1000

#2. BR_2019_data (data is already in mg/L)
#BR_2019_data['DOC avg']  = BR_2019['DOC avg']

#3. PB_2021_data
PB_2021_data['DOC avg'] = (PB_2021_data['DOC avg']*12.01017)/1000

###
# Changing dtypes for Plots 
ARCTIC_ALL['Chlorophyll ug/L']= ARCTIC_ALL['Chlorophyll ug/L'].astype(dtype=np.float64)
ARCTIC_ALL['Sal psu']= ARCTIC_ALL['Sal psu'].astype(dtype=np.float64)
ARCTIC_ALL["Place"] = ARCTIC_ALL["Place"].astype('object')
```
